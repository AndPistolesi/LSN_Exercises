```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x, C, errC = np.loadtxt("ex3.1/out.dat", usecols=(0,8,9), unpack='true')
x1, Cp, errCp = np.loadtxt("ex3.1/out.dat", usecols=(0,2,3), unpack='true')
P, errP = np.loadtxt("ex3.1/out.dat", usecols=(11,12), unpack='true')
Pp, errPp = np.loadtxt("ex3.1/out.dat", usecols=(5,6), unpack='true')

plt.errorbar(x,C,yerr=errC, label='1 passo')
plt.errorbar(x1,Cp,yerr=errCp, label='100 passi')
plt.axhline(y= 14.975790778311286, color='r', linestyle='-')
plt.xlabel('n')
plt.ylabel('option price')
plt.title('Prezzo per una call di un\'opzione europea')

plt.figure()
plt.errorbar(x,P,yerr=errP, label='1 passo')
plt.errorbar(x1,Pp,yerr=errPp, label='100 passi')
plt.axhline(y= 5.4595325819072364, color='r', linestyle='-')
plt.xlabel('n')
plt.ylabel('option price')
plt.title('Prezzo per una put di un\'opzione europea')


plt.show()
```
