```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

points=100
T = np.linspace(0.5,2.0,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )



x, ene, err_ene = np.loadtxt("ISING_1D/ene1.out", usecols=(0,3,4), unpack='true')
x1, ene1, err_ene1 = np.loadtxt("ISING_1D/ene0.out", usecols=(0,3,4), unpack='true')

fig, [ax0, ax1] = plt.subplots(1,2, figsize=(22,8))

ax0.plot(T, e, label="curva teorica")
ax0.errorbar(x,ene,yerr=err_ene, marker="o", markersize='10', linestyle="none", label="simulazione")
ax1.plot(T, e, label="curva teorica")
ax1.errorbar(x1,ene1,yerr=err_ene1, marker="o", markersize='10', linestyle="none", label="simulazione")


ax1.set_title('E(T) usando Gibbs')
ax1.set_ylabel('E')
ax1.set_xlabel('T')

ax0.set_title('E(T) usando Metropolis')
ax0.set_ylabel('E')
ax0.set_xlabel('T')
ax0.legend()
ax1.legend()

plt.show()
```

```python
heat, err_heat = np.loadtxt("ISING_1D/heat1.out", usecols=(3,4), unpack='true')
heat1, err_heat1 = np.loadtxt("ISING_1D/heat0.out", usecols=(3,4), unpack='true')
heat2=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)


fig, [ax0, ax1] = plt.subplots(1,2, figsize=(22,8))


ax0.plot(T, heat2,label="curva teorica")
ax0.errorbar(x,heat,yerr=err_heat, marker="o", markersize='10', linestyle="none", label="simulazione")
ax1.plot(T, heat2, label="curva teorica")
ax1.errorbar(x1,heat1,yerr=err_heat1, marker="o", markersize='10', linestyle="none", label="simulazione")



ax0.set_title('C(T) usando Metropolis')
ax0.set_ylabel('C')
ax0.set_xlabel('T')
ax1.set_title('C(T) usando Gibbs')
ax1.set_ylabel('C')
ax1.set_xlabel('T')
ax0.legend()
ax1.legend()
```

```python
h=0.02
#external field
b = 1/T

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)

mag, err_mag = np.loadtxt("ISING_1D/mag1.out", usecols=(3,4), unpack='true')
mag1, err_mag1 = np.loadtxt("ISING_1D/mag0.out", usecols=(3,4), unpack='true')

fig, [ax0, ax1] = plt.subplots(1,2, figsize=(22,8))

ax0.plot(T, M, label="curva teorica")
ax0.errorbar(x,mag,yerr=err_mag, marker="o", markersize='10', linestyle="none", label="simulazione")
ax1.plot(T, M, label="curva teorica")
ax1.errorbar(x1,mag1,yerr=err_mag1, marker="o", markersize='10', linestyle="none", label="simulazione")


ax1.set_title('M(T) usando Gibbscon h=0.02')
ax1.set_ylabel('M')
ax1.set_xlabel('T')
ax0.set_title('M(T) usando Metropolis con h=0.02')
ax0.set_ylabel('M')
ax0.set_xlabel('T')
ax0.legend()
ax1.legend()
```

```python
chi, err_chi = np.loadtxt("ISING_1D/chi1.out", usecols=(3,4), unpack='true')
chi1, err_chi1 = np.loadtxt("ISING_1D/chi0.out", usecols=(3,4), unpack='true')
X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)

fig, [ax0, ax1] = plt.subplots(1,2, figsize=(22,8))

ax0.plot(T, X, label="curva teorica")
ax0.errorbar(x,chi,yerr=err_chi, marker="o", markersize='10', linestyle="none", label="simulazione")
ax1.plot(T, X, label="curva teorica")
ax1.errorbar(x1,chi1,yerr=err_chi1, marker="o", markersize='10', linestyle="none", label="simulazione")

ax1.set_title('$\chi$(T) usando Metropolis')
ax1.set_ylabel('X')
ax1.set_xlabel('T')
ax0.set_title('$\chi$(T) usando Gibbs')
ax0.set_ylabel('X')
ax0.set_xlabel('T')
ax0.legend()
ax1.legend()
```
